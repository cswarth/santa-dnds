```python
import pandas as pd
from bokeh.plotting import figure, show, ColumnDataSource
from bokeh.io import output_notebook
from bokeh.palettes import Spectral11
from bokeh.models import NumeralTickFormatter
from IPython.display import display
import os
output_notebook()

print(os.getcwd())
```

```python
df = pd.read_csv("build/results.csv")
df.head()
```

```python
from bokeh.models import Jitter
from bokeh.models import HoverTool, BoxSelectTool



factors = list(df.model.unique())
print(factors)
print(df.fitness.unique())

hover = HoverTool(
        tooltips=[
            ("model", "$model"),
            ("fitness", "$fitness"),
            ("generation", "$generation"),
            ("dN/dS", "$omega"),
        ]
    )
TOOLS = [BoxSelectTool(), hover]

fig = figure(title="dN/dS ratios", x_range=factors, 
             width=900, height=450, 
             tools=TOOLS, toolbar_location="above")

fig.title.text_font_size='16pt'
fig.xaxis.axis_label_text_font_size='16pt'
fig.yaxis.axis_label_text_font_size='16pt'

fitness = df.fitness.unique()
colormap = dict(zip(fitness,[Spectral11[i*9] for i in range(len(fitness))]))

for i,(k,grp) in enumerate(df.groupby("model")):
    colors = [colormap[x] for x in grp.fitness]
    fig.circle(x={'value':i+1, 'transform': Jitter(width=0.1)}, y="omega", 
               source=ColumnDataSource(data=grp),
               color=colors, alpha=0.5, size=10)
    
fig.xaxis.axis_label_text_font_size='16pt'
fig.yaxis.axis_label_text_font_size='16pt'

show(fig)
```

```python

from bokeh.plotting import figure, show
from bokeh.models.sources import ColumnDataSource
from bokeh.models.transforms import Jitter
from bokeh.models import Jitter
import pandas as pd
import numpy as np

df = pd.DataFrame.from_dict({
        "model" : np.random.choice(['a', 'b', 'c', 'd'], 50), 
        "omega" : range(50)
        })
factors = list(df.model.unique())
fig = figure(title="jitter plot", x_range=factors)

for i,(k,grp) in enumerate(df.groupby("model")):
    fig.circle(x={'value':i+1, 'transform': Jitter(width=0.4)}, y="omega", 
               source=ColumnDataSource(data=grp),
               alpha=0.3, size=10)

show(fig)
```

```python
from bokeh.plotting import figure, output_file, show, ColumnDataSource
from bokeh.models import HoverTool

output_file("toolbar.html")

source = ColumnDataSource(
        data=dict(
            x=[1, 2, 3, 4, 5],
            y=[2, 5, 8, 2, 7],
            desc=['A', 'b', 'C', 'd', 'E'],
        )
    )

hover = HoverTool(
        tooltips=[
            ("index", "$index"),
            ("(x,y)", "($x, $y)"),
            ("desc", "@desc"),
        ]
    )

p = figure(plot_width=400, plot_height=400, tools=[hover],
           title="Mouse over the dots")

p.circle('x', 'y', size=20, source=source)

show(p)
```

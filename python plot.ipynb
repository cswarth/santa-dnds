```python
import pandas as pd
from bokeh.plotting import figure, show, ColumnDataSource
from bokeh.io import output_notebook
from bokeh.palettes import Spectral11
from bokeh.models import NumeralTickFormatter
from IPython.display import display
import os
output_notebook()

print(os.getcwd())
```

```python
df = pd.read_csv("build/results.csv")
df.head()
```

```python
from bokeh.models import Jitter
from bokeh.models import HoverTool, BoxSelectTool



factors = list(df.model.unique())
print(factors)
print(df.fitness.unique())

hover = HoverTool(
        tooltips=[
            ("model", "$model"),
            ("fitness", "$fitness"),
            ("generation", "$generation"),
            ("dN/dS", "$omega"),
        ]
    )
TOOLS = [BoxSelectTool(), hover]

fig = figure(title="dN/dS ratios - colored according to fitness value", x_range=factors, 
             width=900, height=450, 
             tools=TOOLS, toolbar_location="above")

fig.title.text_font_size='16pt'
fig.xaxis.axis_label_text_font_size='16pt'
fig.yaxis.axis_label_text_font_size='16pt'

fitness = df.fitness.unique()
colormap = dict(zip(fitness,[Spectral11[i*9] for i in range(len(fitness))]))

for i,(k,grp) in enumerate(df.groupby("model")):
    colors = [colormap[x] for x in grp.fitness]
    fig.circle(x={'value':i+1, 'transform': Jitter(width=0.1)}, y="omega", 
               source=ColumnDataSource(data=grp),
               color=colors, alpha=0.5, size=10)
    

show(fig)
```

```python
# example of jitter usage posted to Google groups
# 


from bokeh.plotting import figure, show
from bokeh.layouts import row
from bokeh.models.sources import ColumnDataSource
from bokeh.models.transforms import Jitter
from bokeh.models import Jitter
import pandas as pd
import numpy as np

df = pd.DataFrame.from_dict({
        "model" : np.random.choice(['a', 'b', 'c', 'd'], 50), 
        "omega" : range(50)
        })
factors = list(df.model.unique())

# without jitter.
fig = figure(width=250, plot_height=250, title="no jitter plot", x_range=factors)
fig.toolbar_location = None
fig.circle(x="model", y="omega", source=ColumnDataSource(data=df),
             alpha=0.3, size=5)


# with jitter
jitfig = figure(width=250, plot_height=250, title="jitter plot", x_range=factors)
jitfig.toolbar_location = None
for i,(k,grp) in enumerate(df.groupby("model")):
    print(k)
    jitfig.circle(x={"values":i+1}, y="omega", 
               source=ColumnDataSource(data=grp),
               alpha=0.3, size=5)

p = row(fig, jitfig)

show(p)
```

```python
# example of jitter usage posted to Google groups
# 


from bokeh.plotting import figure, show
from bokeh.layouts import row
from bokeh.models.sources import ColumnDataSource
from bokeh.models.transforms import Jitter
from bokeh.models import Jitter
import pandas as pd
import numpy as np

df = pd.DataFrame.from_dict({
        "model" : np.random.choice(['a', 'b', 'c', 'd'], 50), 
        "omega" : range(50)
        })
factors = list(df.model.unique())

# without jitter.
fig = figure(width=250, plot_height=250, title="no jitter plot", x_range=factors)
fig.toolbar_location = None
c = fig.circle(x="model", y="omega", source=ColumnDataSource(data=df),
             alpha=0.3, size=5)
print(c)
print(c.__dict__)
print(c.__dict__['_property_values']['data_source'])
ds = c.__dict__['_property_values']['data_source']
print(ds.__dict__)

show(fig)
```
